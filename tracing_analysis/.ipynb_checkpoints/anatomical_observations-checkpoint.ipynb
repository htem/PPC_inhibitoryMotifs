```python
# plot neuron morphologies

def plot_neuron(plotDict, ax = None):
    myCnPre = pymaid.get_connectors(plotDict['myNeuron'],relation_type='presynaptic_to')
    if ax is not None:
        pymaid.plot2d(plotDict['myNeuron'],ax=ax,connectors=True,
                linewidth=1.5,method='2d',cn_size=2,color='#414073',scalebar=100)
    else:
        fig,ax = pymaid.plot2d(plotDict['myNeuron'],connectors=True,
                linewidth=1.5,method='2d',cn_size=2,color='#414073',scalebar=100)
    if myAxonTrunk:
        pymaid.plot2d(plotDict['myAxonTrunk'],ax=ax,linewidth=2,color='#588157')
    if myAxonCollaterals:
        pymaid.plot2d(plotDict['myAxonCollaterals'],ax=ax,linewidth=1.5,color='#DAD7CD')
    if myMyelination:
        pymaid.plot2d(plotDict['myMyelination'],ax=ax,linewidth=4,color='black')
    if len(myCnPre)>0:
        pymaid.plot2d(myCnPre[['x','y','z']].values, method='2d', ax=ax,
        scatter_kws={'s': 10,'color': 'r','marker': 'D'})
    return ax
```

```python
def calc_conv_syn_frac(source_df, target_df):
    syn_df = ppc.get_synapses_between(source_df,target_df)
    num_syn = len(syn_df)
    num_conv_syn = len(syn_df)-len(target_df)                            
    conv_perc = (len(syn_df)-len(target_df))/len(syn_df)
    return num_syn, num_conv_syn, conv_perc
    
(num_syn_L23, num_conv_syn_L23, conv_perc_L23) = calc_conv_syn_frac(myNeuronsDf,psp_df[psp_df.layer=='L23'])
(num_syn_L23_pyr, num_conv_syn_L23_pyr, conv_perc_L23_pyr) = calc_conv_syn_frac(myNeuronsDf,psp_df[psp_df.layer=='L23'][psp_df.type=='pyramidal'])
(num_syn_L23_int, num_conv_syn_L23_int, conv_perc_L23_int) = calc_conv_syn_frac(myNeuronsDf,psp_df[psp_df.layer=='L23'][psp_df.type=='non pyramidal'])
                                                                    
(num_syn_L5, num_conv_syn_L5, conv_perc_L5) = calc_conv_syn_frac(myNeuronsDf,psp_df[psp_df.layer=='L5'])
(num_syn_L5_pyr, num_conv_syn_L5_pyr, conv_perc_L5_pyr) = calc_conv_syn_frac(myNeuronsDf,psp_df[psp_df.layer=='L5'][psp_df.type=='pyramidal'])
(num_syn_L5_int, num_conv_syn_L5_int, conv_perc_L5_int) = calc_conv_syn_frac(myNeuronsDf,psp_df[psp_df.layer=='L5'][psp_df.type=='non pyramidal'])
                                    
print('Convergent synapses to L23 cells: %i/%i, %0.2f' % (num_conv_syn_L23, num_syn_L23, conv_perc_L23))
print('Convergent synapses to L23 pyramidal cells: %i/%i, %0.2f' % (num_conv_syn_L23_pyr, num_syn_L23_pyr, conv_perc_L23_pyr))
print('Convergent synapses to L23 non pyramidal cells: %i/%i, %0.2f' % (num_conv_syn_L23_int, num_syn_L23_int, conv_perc_L23_int))

                                   
print('Convergent synapses to L5 cells: %i/%i, %0.2f' % (num_conv_syn_L5, num_syn_L5, conv_perc_L5))
print('Convergent synapses to L5 pyramidal cells: %i/%i, %0.2f' % (num_conv_syn_L5_pyr, num_syn_L5_pyr, conv_perc_L5_pyr))
print('Convergent synapses to L5 non pyramidal cells: %i/%i, %0.2f' % (num_conv_syn_L5_int, num_syn_L5_int, conv_perc_L5_int))
```

```python
def plot_DiGraph(DiG, weight='psd_area', node_color=None, node_size=None):
    #fig, ax = plt.subplots(figsize=(20,20))
    edge_widths = np.array([i[weight] for i in dict(DiG.edges).values()])
    # normalize edge widths
    edge_widths = edge_widths/np.mean(edge_widths)*1
    #pos = nx.spring_layout(DiG,weight=weight,iterations=100)
    pos = nx.kamada_kawai_layout(DiG,weight=weight)
    nx.draw_networkx_nodes(DiG, pos, labels=True, node_color=node_color, node_size=node_size)
    nx.draw_networkx_edges(DiG, pos, width=edge_widths)

DiG = ppc.DiGraph_from_syn(synapse_df)
node_color = ['cyan' if str(i) in mySources.skeleton_id.values else 'gray' for i in DiG.nodes]  
node_size = [300 if str(i) in mySources.skeleton_id.values else 20 for i in DiG.nodes]
fig, ax = plt.subplots(figsize=(10,10))
plot_DiGraph(DiG, 'psd_area', node_color=node_color, node_size=node_size)
```

```python
# Calculate soma-to-syapse distances (slow)
def add_axon_syn_dists(synapse_df):
    def get_syn_dist(row):
        connector_id = row.connector_id
        myNeuron = pymaid.get_neuron(row.source)
        presynaptic_to_node = pymaid.get_connector_details(connector_id).presynaptic_to_node
        syn_to_soma_dist = pymaid.dist_between(myNeuron, presynaptic_to_node, myNeuron.soma)/1000
        
        return syn_to_soma_dist
            
    synapse_df['syn_dist_to_soma'] = synapse_df.apply (lambda row:
        get_syn_dist(row), axis=1)
    return synapse_df

pymaid.clear_cache()
synapse_df = add_axon_syn_dists(synapse_df)
```
