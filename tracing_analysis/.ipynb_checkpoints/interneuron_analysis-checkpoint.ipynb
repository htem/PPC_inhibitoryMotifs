```python
import pymaid
import matplotlib.pyplot as plt
import numpy as np
import tracing_analysis_functions_PPC as ppc
import importlib
importlib.reload(ppc)

import seaborn as sns
import pandas as pd
import sys
from datetime import date
import time
from scipy.spatial import distance as dist
import scipy.io as sio
import pickle
import networkx as nx
from sklearn.linear_model import LinearRegression

# Supress default INFO logging
import logging, sys
logging.disable(sys.maxsize)

rm = pymaid.CatmaidInstance('http://catmaid3.hms.harvard.edu/catmaidppc',
                            'akt13',	
                            'Tor5ion25',
                            '9afd2769efa5374b8d48cb5c52af75218784e1ff')

labels = pymaid.get_label_list()
mySessions = ['LD187_141216','LD187_141215','LD187_141214','LD187_141213','LD187_141212',
             'LD187_141211','LD187_141210','LD187_141209','LD187_141208','LD187_141207','LD187_141206']

def init_neuronDf(skelID, labels, isMatched=False):
    #print('Initializing Skel %s' % skelID)
    myNeuronDf = pd.Series()
    myLabels = labels[ labels.skeleton_id == int(skelID) ]
    myNeuron = pymaid.get_neuron(skelID)
    myNeuronDf = ppc.query_cell_info(myNeuronDf,myNeuron,myLabels)

    if isMatched:
        myCellID = ppc.get_cid_from_skelID(skelID)
        myNeuronDf['matched_cell_ID'] = myCellID
        myNeuronDf['corr_quality'] = ppc.get_corr_quality(skelID)
        ppc.get_2P_ROIs(myNeuronDf,myCellID,myNeuron,myLabels,mySessions)
        ppc.query_RL_selectivity(myNeuronDf,myCellID,myNeuron,myLabels,mySessions,RL_selectIdx)
        ppc.query_psp_tracing(myNeuronDf,myCellID,myNeuron,myLabels)
    #ppc.query_axon(myNeuronDf,myCellID,myNeuron,myLabels)
    #ppc.query_BS_inputs(myNeuronDf,myCellID,myNeuron,myLabels)
    #ppc.query_CC_inputs(myNeuronDf,myCellID,myNeuron,myLabels)
    return(myNeuronDf)


def add_axon_overlap(source_df, psp_df, labels, max_dist=20):
    def get_axon_overlap(row, labels, max_dist):
        targetNeuron = pymaid.get_neuron(row['skeleton_id'])
        try:
        #if True:
            myLabels = labels[ labels.skeleton_id == int(targetNeuron.skeleton_id) ]
            mySoma = get_soma_only(targetNeuron, myLabels)
            axon_overlap = pymaid.cable_overlap(mySourceAxon, mySoma, dist=max_dist, method='max')/1000 #units in microns
            return axon_overlap.values[0][0]
        except:
            print('Error loading skel_id %s' % targetNeuron.skeleton_id)
            return np.nan # This isnt great, should revisit      
    
    sourceNeuron = pymaid.get_neuron(source_df.skeleton_id)
    # Define axon based on "axon" tag and count synapses
    myLabels = labels[ labels.skeleton_id == int(sourceNeuron.skeleton_id) ]
    myAxnLbl = myLabels[ myLabels.tag == 'axon']

    if len(myAxnLbl) < 1:
        print('WARNING: no axon tag on skel %s' % sourceNeuron.skeleton_id)
        return (None, None, None, None)
    elif len(myAxnLbl) == 1:
        myAxnStartID = myAxnLbl.treenode_id.values
    else:
        print('WARNING: more than one axon tag on skel %s' % sourceNeuron.skeleton_id)
        myAxnStartID = myAxnLbl.treenode_id.values[0]

    myAxon = pymaid.cut_neuron(sourceNeuron,int(myAxnStartID),ret='distal')
    mySourceAxon = myAxon
    
    
    psp_df['axon_overlap'] = psp_df.apply (lambda row:
        get_axon_overlap(row, labels, max_dist), axis=1)
    return psp_df


def add_connection_weights(source_df, psp_df, synapse_df):
    def get_connection_weight(source_skid, target_skid, DiG, weight):
        if (source_skid, target_skid) in dict(DiG.edges).keys():
            return dict(DiG.edges)[(source_skid, target_skid)][weight]
        else: 
            return 0
    
    
    DiG = ppc.DiGraph_from_syn(synapse_df)
    weight='count'
    
    source_skid = int(source_df.skeleton_id)

    psp_df['syn_count']=psp_df.apply (lambda row: get_connection_weight(source_skid, int(row.skeleton_id), DiG, 'count'), axis=1)
    psp_df['psd_area']=psp_df.apply (lambda row: get_connection_weight(source_skid, int(row.skeleton_id), DiG, 'psd_area'), axis=1)
   
    return psp_df

def get_soma_only(myNeuron, myLabels):
    
    mySomaLbl = myLabels[ myLabels.tag == 'soma']
    if len(mySomaLbl) < 1:
        print('WARNING: no soma tag on skel %s' % myNeuron.skeleton_id)
        return None
    elif len(mySomaLbl) == 1:
        mySomaID = mySomaLbl.treenode_id.values[0]
    else:
        print('WARNING: more than soma tag on skel %s' % myNeuron.skeleton_id)
        mySomaID = mySomaLbl.treenode_id.values[0]
        
    # Hack - neuron needs more than 1 node. Add all direct children of soma
    #print(mySomaID)
    include_nodes = list(myNeuron.graph.predecessors(mySomaID))
    include_nodes.append(mySomaID)
    mySoma = pymaid.subset_neuron(myNeuron,include_nodes)
    return mySoma

def calc_axon_soma_dists(source_df, targets_df, labels, max_dist=50):

    sourceNeuron = pymaid.get_neurons(source_df.skeleton_id)
    targetNeurons = pymaid.get_neurons(targets_df.skeleton_id)
    
    
    # Define axon based on "axon" tag and count synapses
    myLabels = labels[ labels.skeleton_id == int(sourceNeuron.skeleton_id) ]
    myAxnLbl = myLabels[ myLabels.tag == 'axon']

    if len(myAxnLbl) < 1:
        print('WARNING: no axon tag on skel %s' % sourceNeuron.skeleton_id)
        return (None, None, None)
    elif len(myAxnLbl) == 1:
        myAxnStartID = myAxnLbl.treenode_id.values
    else:
        print('WARNING: more than one axon tag on skel %s' % sourceNeuron.skeleton_id)
        myAxnStartID = myAxnLbl.treenode_id.values[0]

    myAxon = pymaid.cut_neuron(sourceNeuron,int(myAxnStartID),ret='distal')
    mySourceAxon = myAxon
    
    (axon_overlaps, somasList) = (list(), list())

    for idx,targetNeuron in enumerate(targetNeurons):
        try:
            myLabels = labels[ labels.skeleton_id == int(targetNeuron.skeleton_id) ]
            mySoma = get_soma_only(targetNeuron, myLabels)
            axon_overlap = pymaid.cable_overlap(mySourceAxon, mySoma, dist=max_dist, method='max')/1000 #units in microns
            axon_overlaps.append(axon_overlap.values[0][0])
            somasList.append(mySoma)
        except:
            print('Error loading skel_id %s' % targetNeuron.skeleton_id)
            axon_overlaps.append(np.nan)
            somasList.append(targetNeuron)
    mySomas = pymaid.core.CatmaidNeuronList(somasList)
    
    return (myAxon, mySomas, np.array(axon_overlaps))
```

```python
# Plot synapse count vs axon overlap and fit model

def scatter_and_fit(X, y, xlabel=None, ylabel=None, plot=True, fit_intercept = True):
    from sklearn.linear_model import LinearRegression
    X_ = X.reshape(-1, 1)
    y_ = y.reshape(-1, 1)
    reg = LinearRegression(fit_intercept=fit_intercept).fit(X_, y_)
    y_pred = reg.predict(X_)

    if plot == True:
        #sns.scatterplot(X, y);
        if xlabel is not None:
            plt.xlabel(xlabel)
        if ylabel is not None:
            plt.ylabel(ylabel)
        plt.plot(X,y_pred,'gray')
        plt.title('R2 = %0.2f' % reg.score(X_,y_))
    return reg, reg.score(X_,y_)


sns.set(rc={'figure.figsize':(8,4)}, font_scale=2)

MN_and_PSP_df = pd.concat([MN_df, myPSPs_df])
MN_and_PSP_df=MN_and_PSP_df[MN_and_PSP_df.axon_overlap<15] # remove 2 outliers
#sns.scatterplot(x='axon_overlap', y='syn_count', data=MN_and_PSP_df)

mask = MN_and_PSP_df.axon_overlap.values >= 0
(reg, r2)= scatter_and_fit(MN_and_PSP_df.axon_overlap.values[mask], 
                    MN_and_PSP_df.syn_count.values[mask], plot=True, fit_intercept=True)
g = sns.scatterplot(x='axon_overlap', y='syn_count',data=MN_and_PSP_df, size_norm = 100)

g = sns.scatterplot(x='axon_overlap', y='syn_count',data=myMN_PSPs_df, size_norm = 100, hue='RL_select_idx', hue_norm=(-.5, .5),palette='RdBu')
g.legend(loc='center left', bbox_to_anchor=(1.25, 0.5), ncol=1)
```

```python
def calc_morphology_overlap(myNeuronsDf, labels, max_dist):

    soma_locs = np.array([i for i in myNeuronsDf.soma_loc.values])#[:,:,0]
    soma_dists = dist.pdist(soma_locs)/1000
    soma_dists_xy = dist.pdist(soma_locs[:,[0,1]])/1000
    soma_dists_xz = dist.pdist(soma_locs[:,[0,2]])/1000
    soma_dists_yz = dist.pdist(soma_locs[:,[1,2]])/1000
    soma_dists_x = dist.pdist(soma_locs[:,[0]])/1000

    myNeurons = pymaid.get_neurons(myNeuronsDf[:].skeleton_id)
    (axonList, trunkList, collatList) = (list(), list(), list())

    for idx,myNeuron in enumerate(myNeurons):
        myLabels = labels[ labels.skeleton_id == int(myNeuron.skeleton_id) ]
        (myAxon, myAxonTrunk, myAxonCollaterals, myMyelination) = get_axon_components(myNeuron, myLabels);
        axonList.append(myAxon)
        trunkList.append(myAxonTrunk)
        collatList.append(myAxonCollaterals)
    myAxons = pymaid.core.CatmaidNeuronList(axonList)
    myAxonTrunks = pymaid.core.CatmaidNeuronList(trunkList)
    myAxonCollats = pymaid.core.CatmaidNeuronList(collatList)

    axon_overlap = dist.squareform(pymaid.cable_overlap(myAxons, myAxons, dist=max_dist, method='min'), checks=False)
    trunk_overlap = dist.squareform(pymaid.cable_overlap(myAxonTrunks, myAxonTrunks, dist=max_dist, method='min'), checks=False)
    collat_overlap = dist.squareform(pymaid.cable_overlap(myAxonCollats, myAxonCollats, dist=max_dist, method='min'), checks=False)

    return (soma_dists, axon_overlap, trunk_overlap, collat_overlap)

def get_axon_components(myNeuron, myLabels):
    # Define axon based on "axon" tag and count synapses
    myAxnLbl = myLabels[ myLabels.tag == 'axon']

    if len(myAxnLbl) < 1:
        print('WARNING: no axon tag on skel %s' % myNeuron.skeleton_id)
        return (None, None, None, None)
    elif len(myAxnLbl) == 1:
        myAxnStartID = myAxnLbl.treenode_id.values
    else:
        print('WARNING: more than one axon tag on skel %s' % myNeuron.skeleton_id)
        myAxnStartID = myAxnLbl.treenode_id.values[0]

    myAxon = pymaid.cut_neuron(myNeuron,int(myAxnStartID),ret='distal')

    # Separate axon trunk
    axonTrunkNodes = np.asarray(myAxon.nodes.treenode_id[myAxon.nodes.radius.values>0])
    axonCollateralNodes = np.asarray(myAxon.nodes.treenode_id[myAxon.nodes.radius.values<0])
    myAxonTrunk = pymaid.subset_neuron(myAxon,axonTrunkNodes)
    myAxonCollaterals = pymaid.subset_neuron(myAxon,axonCollateralNodes)

    myMyn = myLabels[ myLabels.tag == 'myelinates']
    myUnmyn = myLabels[ myLabels.tag == 'unmyelinates']
    myMynNodeIDs = find_myelination(myNeuron,myMyn,myUnmyn)
    myMyelination = pymaid.subset_neuron(myNeuron,np.asarray(myMynNodeIDs))

    return myAxon, myAxonTrunk, myAxonCollaterals, myMyelination
```

```python
def get_cid_from_skelID(skelID):
    cid = np.nan
    annotations = pymaid.get_annotations(skelID)
    #annotations[str(skelID)]
    annotTag = 'new matched neuron '
    if str(skelID) in annotations.keys():
        result = [i for i in annotations[str(skelID)] if i.startswith(annotTag)]
        if len(result) > 0:
            if len(result) > 1:
                print('WARNING: more than one matched neuron num')
            #cid = result[0][-4:]
            try:
                cid = float(result[0][len(annotTag):])
            except:
                print('Warning - non numerical cid')    
    return cid
```
