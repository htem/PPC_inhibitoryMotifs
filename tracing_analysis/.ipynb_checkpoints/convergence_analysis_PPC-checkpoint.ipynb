```python
import pymaid
import matplotlib.pyplot as plt
import numpy as np
import tracing_analysis_functions_PPC as ppc
import importlib
importlib.reload(ppc)
import seaborn as sns
import pandas as pd
import sys
import os
import h5py
from datetime import date
import time
from scipy.spatial import distance as dist
import scipy.io as sio
import pickle
import networkx as nx
from sklearn.linear_model import LinearRegression
import scipy.stats as stats


# Supress default INFO logging
import logging, sys
logging.disable(sys.maxsize)

#rm = pymaid.CatmaidInstance('http://catmaid3.hms.harvard.edu/catmaidppc',
#                            'akt13',	
#                            'Tor5ion25',
#                            '9afd2769efa5374b8d48cb5c52af75218784e1ff')

rm = pymaid.CatmaidInstance('http://catmaid3.hms.harvard.edu/catmaidppc',
                            api_token='9afd2769efa5374b8d48cb5c52af75218784e1ff',
                            http_user='akt13')#, # omit if not required
                            #http_password='pw') # omit if not required

labels = pymaid.get_label_list()
mySessions = ['LD187_141216','LD187_141215','LD187_141214','LD187_141213','LD187_141212','LD187_141211',
              'LD187_141210','LD187_141209','LD187_141208','LD187_141207','LD187_141206']

def init_neuronDf(skelID, labels, isMatched=False):
    #print('Initializing Skel %s' % skelID)
    myNeuronDf = pd.Series()
    myLabels = labels[ labels.skeleton_id == int(skelID) ]
    myNeuron = pymaid.get_neuron(skelID)
    myNeuronDf = ppc.query_cell_info(myNeuronDf,myNeuron,myLabels)

    if isMatched:
        myCellID = ppc.get_cid_from_skelID(skelID)
        myNeuronDf['matched_cell_ID'] = myCellID
        myNeuronDf['corr_quality'] = ppc.get_corr_quality(skelID)
        ppc.get_2P_ROIs(myNeuronDf,myCellID,myNeuron,myLabels,mySessions)
        ppc.query_RL_selectivity(myNeuronDf,myCellID,myNeuron,myLabels,mySessions,RL_selectIdx)
        ppc.query_ROC_selectivity(myNeuronDf,myCellID,myNeuron,myLabels,mySessions,AUC_fullTrial,
            RL_selectivity_ROC,RL_selectivity_ROC_blocks)
        ppc.query_psp_tracing(myNeuronDf,myCellID,myNeuron,myLabels)
        ppc.query_trial_snr(myNeuronDf,myCellID,myNeuron,myLabels,mySessions,trial_snr)
    #ppc.query_axon(myNeuronDf,myCellID,myNeuron,myLabels)
    #ppc.query_BS_inputs(myNeuronDf,myCellID,myNeuron,myLabels)
    #ppc.query_CC_inputs(myNeuronDf,myCellID,myNeuron,myLabels)
    return(myNeuronDf)

def source_target_labels(myNeuronsDf, field = 'RL_selectivity'):
    skids = myNeuronsDf[field].values
    source_mat = np.tile(skids,(len(skids),1))
    target_mat = np.tile(skids,(len(skids),1)).transpose()
    source_cond = dist.squareform(source_mat, checks=False)
    target_cond = dist.squareform(target_mat, checks=False)
    return (source_cond, target_cond)

def pair_labels(myNeuronsDf):
    skids = myNeuronsDf.matched_cell_ID.values.astype(int)
    source_mat = np.tile(skids,(len(skids),1))
    target_mat = np.tile(skids,(len(skids),1)).transpose()
    source_cond = dist.squareform(source_mat, checks=False)
    target_cond = dist.squareform(target_mat, checks=False)
    return (source_cond, target_cond)

def pair_select(myNeuronsDf, select_idx = 'RL_select_idx'):
    skids = myNeuronsDf[select_idx].values
    source_mat = np.tile(skids,(len(skids),1))
    target_mat = np.tile(skids,(len(skids),1)).transpose()
    source_cond = dist.squareform(source_mat, checks=False)
    target_cond = dist.squareform(target_mat, checks=False)
    return (source_cond, target_cond)

def pair_trial_snr(myNeuronsDf):
    skids = myNeuronsDf.trial_snr_max.values
    source_mat = np.tile(skids,(len(skids),1))
    target_mat = np.tile(skids,(len(skids),1)).transpose()
    source_cond = dist.squareform(source_mat, checks=False)
    target_cond = dist.squareform(target_mat, checks=False)
    return (source_cond, target_cond)

def plt_corr_vs_sim(corr, sim, xlabel=None, ylabel=None, filter_dists = None,neuron_labels = None):
    if filter_dists is not None:
        corr = corr[filter_dists]
        sim = sim[filter_dists]
    sns.scatterplot(corr, sim)    
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    if neuron_labels is not None:        
        (source_labels, target_labels) = pair_labels(neuron_labels)
        if filter_dists is not None:
            source_labels = source_labels[filter_dists]
            target_labels = target_labels[filter_dists]
        for idx in np.arange(len(source_labels)): 
            plt.text(corr[idx],sim[idx],(source_labels[idx],target_labels[idx]), horizontalalignment='left', 
            size='large', color='black', weight='semibold')

def scatter_and_fit(X, y, xlabel=None, ylabel=None, plot=True):
    X_ = X.reshape(-1, 1)
    y_ = y.reshape(-1, 1)
    reg = LinearRegression().fit(X_, y_)
    y_pred = reg.predict(X_)
    
    if plot == True:
        sns.scatterplot(X, y);
        if xlabel is not None:
            plt.xlabel(xlabel)
        if ylabel is not None:
            plt.ylabel(ylabel)
        plt.plot(X,y_pred)
        plt.title('R2 = %0.2f' % reg.score(X_,y_))
    return reg.score(X_,y_)
```

```python
# generate pair types df (with shuffle)
def calc_conv_pairs(mySources, myCN_df):
    def pair_diff(myNeuronsDf, col='RL_select_idx_ROC'):
        cols = myNeuronsDf[col].values.astype(float)
        source_mat = np.tile(cols,(len(cols),1))
        target_mat = np.tile(cols,(len(cols),1)).transpose()
        diff_mat = np.absolute(source_mat - target_mat)
        diff_mat = dist.squareform(diff_mat, checks=False)
        return (diff_mat)
    def calc_RL_pair_types(myNeuronsDf,selectivity_type = 'RL_selectivity'):
        num_cells = len(myNeuronsDf)
        RL_pair_types =  [[''] * num_cells for i in range(num_cells)]
        RL_pair_types_detail =  [[''] * num_cells for i in range(num_cells)]

        for idx_1,RL_1 in enumerate(myNeuronsDf[selectivity_type]):
            for idx_2, RL_2 in enumerate(myNeuronsDf[selectivity_type]):
                if RL_1 != 'Non-selective' and RL_1 != 'Mixed' and RL_2 != 'Non-selective' and RL_2 !='Mixed':
                    if RL_1 == RL_2:
                        RL_pair_types[idx_1][idx_2] = 'Same'
                        if RL_1 == 'Left':
                            RL_pair_types_detail[idx_1][idx_2] = 'Left - Left'
                        if RL_1 == 'Right':
                            RL_pair_types_detail[idx_1][idx_2] = 'Right - Right'
                    else:
                        RL_pair_types[idx_1][idx_2] = 'Different'
                        RL_pair_types_detail[idx_1][idx_2] = 'Right - Left'
                elif RL_1 == 'Non-selective' and RL_2 == 'Non-selective':
                    RL_pair_types[idx_1][idx_2] = 'Same'
                    RL_pair_types_detail[idx_1][idx_2] = 'Non - Non'
                else:
                    RL_pair_types[idx_1][idx_2] = 'Different'
                    if RL_1 == 'Right' or RL_2 == 'Right':
                        RL_pair_types_detail[idx_1][idx_2] = 'Non - Right'
                    else:
                        RL_pair_types_detail[idx_1][idx_2] = 'Non - Left'
        return dist.squareform(RL_pair_types,checks=False), dist.squareform(RL_pair_types_detail,checks=False)

    def calc_pair_metrics(mySources, myCN_df):   
        # calculate activity and connectivity similarity metrics
        myDiff_ROC = pair_diff(mySources, col='RL_select_idx_ROC')
        myDiff_RL = pair_diff(mySources, col='RL_select_idx')

        (avg_conv_psd_area_sq, avg_conv_psd_area) = ppc.get_cn_sim(mySources.skeleton_id, synapse_df=myCN_df, metric='avg_conv_psd_area',weight='avg_psd_area')
        avg_conv_psd_area[avg_conv_psd_area==0]=np.nan

        # calc selectivity categories
        (source_selectivity, target_selectivity) = source_target_labels(mySources, field='RL_selectivity')
        (source_selectivity_ROC, target_selectivity_ROC) = source_target_labels(mySources, field='RL_selectivity_ROC')

        # neuron labels
        (source_labels, target_labels) = pair_labels(mySources)

        # calc activity correlations
        (myCorr_trial_avg_sq, myCorr_trial_avg, sessions_overlap) = ppc.calc_act_corr(mySources, corr_trial_avg, mySessions)

        # generate pair metrics
        (my_RL_pair_types, my_RL_pair_types_detail) = calc_RL_pair_types(mySources,  selectivity_type='RL_selectivity')
        (my_ROC_RL_pair_types, my_ROC_RL_pair_types_detail) = calc_RL_pair_types(mySources, selectivity_type='RL_selectivity_ROC')
        
        # generate pair metrics for individual blocks
        block_labels = ('cueEarly','cueLate','delayEarly','delayTurn','turnITI')
        block_pair_types = dict()
        for block in block_labels:
            (block_pair_types[block],detail) =  calc_RL_pair_types(mySources, selectivity_type=block)            
            
        pair_types_df = dict()
        for (i,source) in enumerate(source_labels):
            pair_types_df['source_selectivity'] = source_selectivity
            pair_types_df['target_selectivity'] = target_selectivity
            pair_types_df['RL_diff_RL'] = myDiff_RL

            pair_types_df['source_selectivity_ROC'] = source_selectivity_ROC
            pair_types_df['target_selectivity_ROC'] = target_selectivity_ROC
            pair_types_df['RL_diff_ROC'] = myDiff_ROC

            pair_types_df['avg_conv_psd_area'] = avg_conv_psd_area

            pair_types_df['pair_selectivity'] = my_RL_pair_types
            pair_types_df['pair_selectivity_detail'] = my_RL_pair_types_detail
            pair_types_df['pair_selectivity_ROC'] = my_ROC_RL_pair_types
            pair_types_df['pair_selectivity_ROC_detail'] = my_ROC_RL_pair_types_detail
            
            for block in block_labels:
                pair_types_df[block] = block_pair_types[block]

            pair_types_df['source'] = source_labels
            pair_types_df['target'] = target_labels

            pair_types_df['corr_trial_avg'] = myCorr_trial_avg 
        return pair_types_df

    pair_types = calc_pair_metrics(mySources, myCN_df)
    return pair_types

```

```python
def pair_diff(myNeuronsDf, col='RL_select_idx_ROC'):
        cols = myNeuronsDf[col].values.astype(float)
        source_mat = np.tile(cols,(len(cols),1))
        target_mat = np.tile(cols,(len(cols),1)).transpose()
        diff_mat = np.absolute(source_mat - target_mat)
        diff_mat = dist.squareform(diff_mat, checks=False)
        return (diff_mat)
sns.set(rc={'figure.figsize':(6,6)})
sns.scatterplot(pair_diff(mySources, col='RL_select_idx_ROC'), avg_conv_psd_area)
```

```python
def calc_act_corr(myNeuronsDf, corr_aligned, mySessions):

    corr_mat_combined = np.empty([len(myNeuronsDf), len(myNeuronsDf), len(mySessions)])
    sessions_overlap = np.zeros([len(myNeuronsDf), len(myNeuronsDf)])
    corr_mat_combined[:] = np.nan
    for idx_1,id_1 in enumerate(myNeuronsDf.matched_cell_ID):
        neuron_1 = myNeuronsDf[myNeuronsDf.matched_cell_ID == id_1]
        for idx_2, id_2 in enumerate(myNeuronsDf.matched_cell_ID):
            neuron_2 = myNeuronsDf[myNeuronsDf.matched_cell_ID == id_2]
            sessions_1_bool = np.array([x>-1 for x in neuron_1.sessions_ROI])
            sessions_2_bool = np.array([x>-1 for x in neuron_2.sessions_ROI])
            match_bool = np.logical_and(sessions_1_bool, sessions_2_bool)
            for session_idx,match in enumerate(match_bool[0]):
                if match:
                    roi_1 = int(neuron_1.sessions_ROI.values[0][session_idx]) # This hacky indexing needs to be fixed when i have time
                    roi_2 = int(neuron_2.sessions_ROI.values[0][session_idx])
                    if idx_1 != idx_2:
                        corr_mat_combined[idx_1, idx_2, session_idx] = corr_aligned[mySessions[session_idx]][roi_1,roi_2]
                        sessions_overlap[idx_1,idx_2] += 1;
    corr_mat_average = np.nanmean(corr_mat_combined,2)
    return corr_mat_average, dist.squareform(corr_mat_average,checks=False), sessions_overlap
```

```python
# Plot convergence analysis as function
def convergence_analysis(myNeuronsDf, myTargets, corr_bins, 
                         weight = None, filter_dists=None, act_corr='aligned', num_shuf=None, plot=False):

    mySources = myNeuronsDf.skeleton_id.values
    G = ppc.get_graph(mySkids, myTargets)

    (shared_cn_sq, shared_cn) = ppc.get_shared_partners(mySources, myTargets)
    if filter_dists is not None:
        shared_cn = shared_cn[filter_dists]
        
    (j_sim_sq, j_sim) = ppc.get_jaccard_sim(mySources, myTargets, weight=weight)
    if filter_dists is not None:
        j_sim = j_sim[filter_dists]
    
    if act_corr=='aligned':
        (myCorr_sq, myCorr) = ppc.calc_trial_aligned_corr(myNeuronsDf, corr_aligned, mySessions)

    elif act_corr=='raw':
        (myCorr_sq, myCorr) = ppc.calc_corr_mat(myNeuronsDf, corr_all, mySessions)
        
    if filter_dists is not None:
        myCorr = myCorr[filter_dists]
    
    corr_bin_labels = np.arange(0,len(corr_bins))+1
    corr_binned_df = pd.DataFrame(columns=['j_sim','perc_conv','data_type','bin'])

    t = time.time()
    for idx,edge in enumerate(corr_bins[:-1]):
        mask = np.logical_and(myCorr > corr_bins[idx],myCorr <= corr_bins[idx+1])
        print('Found %i points ' % sum(mask))
        xlabel = corr_bin_labels[idx]
        j_sim_avg = np.mean(j_sim[mask])
        print(j_sim_avg)
        perc_conv = sum(j_sim[mask] > 0)/sum(mask)
        corr_binned_df = corr_binned_df.append({'j_sim' : j_sim_avg, 'perc_conv' : perc_conv,
            'data_type' : 'data' , 'bin' : xlabel} , ignore_index=True)
        
        if num_shuf is not None:
            for j in np.arange(num_shuf):
                # shuffle the cell's physiology (equiv to shuffling cell identities)
                myNeuronsShuf = myNeuronsDf.copy()
                myNeuronsShuf['sessions_ROI'] = np.random.permutation(myNeuronsDf['sessions_ROI'].values)
                if act_corr=='aligned':
                    (myCorrShuf_sq, myCorrShuf) = ppc.calc_trial_aligned_corr(myNeuronsShuf, corr_aligned, mySessions)                
                elif act_corr=='raw':
                    (myCorrShuf_sq, myCorrShuf) = ppc.calc_corr_mat(myNeuronsShuf, corr_all, mySessions)
                if filter_dists is not None:
                    myCorrShuf = myCorrShuf[filter_dists]
                mask_shuf = np.logical_and(myCorrShuf > corr_bins[idx],myCorrShuf <= corr_bins[idx+1])
                j_sim_shuf = np.mean(j_sim[mask_shuf])
                perc_conv_shuf = sum(j_sim[mask_shuf] > 0)/sum(mask_shuf)
                corr_binned_df = corr_binned_df.append({'j_sim' : j_sim_shuf, 'perc_conv' : perc_conv_shuf,
                    'data_type' : 'shuffle' , 'bin' : xlabel} , ignore_index=True)
                
                
        if plot:
            sns.set(rc={'figure.figsize':(15,5)})
            plt.subplot(1,3,1)
            ax = sns.scatterplot(myCorr,j_sim)
            for x in corr_bins:
                plt.axvline(x=x, color='k', linestyle='-')
            plt.subplot(1,3,2)
            ax=sns.barplot(data = corr_binned_df, y='j_sim',x='bin',hue='data_type', ci = 'sd')
            plt.subplot(1,3,3)
            ax=sns.barplot(data = corr_binned_df, y='perc_conv',x='bin',hue='data_type', ci = 'sd')
    return corr_binned_df
```

```python
# Working code for approx synapse size 

def calc_syn_approx_D(connector_id):
    from scipy.spatial.distance import euclidean   
    cn_details = pymaid.get_connector_details(connector_id)
    pre_s_node = cn_details.presynaptic_to_node
    post_s_node = cn_details.postsynaptic_to_node
    pre_s_node_loc = pymaid.get_node_location(pre_s_node)
    post_s_node_loc = pymaid.get_node_location(post_s_node[0])  
    syn_D = euclidean(pre_s_node_loc[['x','y','z']],post_s_node_loc[['x','y','z']])
    return syn_D

def get_synapses_between(source_skids, target_skids):
    connectors = pymaid.get_connectors_between(source_skids,target_skids)
    synapse_df = connectors[['source_neuron','target_neuron','connector_id']]
    synapse_df['psd_area'] = synapse_df.apply (lambda row: 
        3.14159*(calc_syn_approx_D(row['connector_id'])/2)**2, axis=1)
    synapse_df["count"] = 1    
    synapse_df = synapse_df.rename(columns={"source_neuron": "source", "target_neuron": "target"})
    return synapse_df

def DiGraph_from_syn(synapse_df):
    DiG = nx.from_pandas_edgelist(synapse_df.groupby(['source', 'target'], as_index=False).agg('sum'), 
            edge_attr=['psd_area','count'], create_using=nx.DiGraph)
    return DiG

def plot_DiGraph(DiG, weight='psd_area'):
    fig, ax = plt.subplots(figsize=(5,5))
    edge_widths = np.array([i[weight] for i in dict(G.edges).values()])
       
    # normalize edge widths
    edge_widths = edge_widths/np.mean(edge_widths)*5
    #edge_widths = np.interp(edge_widths, (edge_widths.min(), edge_widths.max()), (1, 10))
    #plot(DiGraph(DiG))

    pos = nx.spring_layout(DiG,weight=weight)

    nx.draw_networkx_nodes(DiG, pos, labels=True)
    nx.draw_networkx_edges(DiG, pos, width=edge_widths)
    
    
synapse_df = get_synapses_between([142171], [165860,165845])
DiG = DiGraph_from_syn(synapse_df)
plot_DiGraph(DiG, 'count')

```

```python

# Calculate weigthed jaccard similarity (aka Ruzicka similarity)
# https://en.wikipedia.org/wiki/Jaccard_index#Weighted_Jaccard_similarity_and_distance

def jaccard_coefficient(G, ebunch=None):
    def predict(u, v):
        union_size = len(set(G[u]) | set(G[v]))
        if union_size == 0:
            return 0
        return len(list(nx.common_neighbors(G, u, v))) / union_size
    return _apply_prediction(G, predict, ebunch)

def _apply_prediction(G, func, ebunch=None):
    """Applies the given function to each edge in the specified iterable
    of edges.

    `G` is an instance of :class:`networkx.Graph`.

    `func` is a function on two inputs, each of which is a node in the
    graph. The function can return anything, but it should return a
    value representing a prediction of the likelihood of a "link"
    joining the two nodes.

    `ebunch` is an iterable of pairs of nodes. If not specified, all
    non-edges in the graph `G` will be used.

    """
    if ebunch is None:
        ebunch = nx.non_edges(G)
    return ((u, v, func(u, v)) for u, v in ebunch)

def jaccard_coefficient_weighted(G, ebunch=None, weight='count'):
    other_nodes = set(G.nodes)
    other_nodes.remove(u)
    other_nodes.remove(v)
    
    def predict(u, v):
        u_edges = [G[u][i][weight] if i in G[u] else 0 for i in other_nodes]
        v_edges = [G[v][i][weight] if i in G[v] else 0 for i in other_nodes]

        numer = 0
        denom = 0

        for idx,node in enumerate(other_nodes):
            numer+= np.minimum(u_edges[idx], v_edges[idx])
            denom+= np.maximum(u_edges[idx], v_edges[idx])

        if denom == 0:
            return 0
        else:
            return numer / denom
    return _apply_prediction(G, predict, ebunch)

def get_jaccard_sim(source_skids, target_skids, weight=None):
    synapse_df = get_synapses_between(source_skids, target_skids)
    DiG = DiGraph_from_syn(synapse_df)
    G = DiG.to_undirected()
    if weight is not None:
        j = jaccard_coefficient_weighted(G, weight=weight)
    else:
        j = nx.jaccard_coefficient(G)
    j_df = pd.DataFrame(j)
    j_df = j_df.rename(columns={0: "source", 1: "target", 2: "weight"})
    j_G = nx.from_pandas_edgelist(j_df, edge_attr='weight')
    j_sq = nx.to_pandas_adjacency(j_G, nodelist = np.array(source_skids,dtype=int))
    
    return j_sq, dist.squareform(j_sq,checks=False)

fig, ax = plt.subplots(figsize=(10,5))
plt.subplot(1,2,1)
(j_sq, j) = get_jaccard_sim(convMNlist.skeleton_id, convPSpartnersList.skeleton_id)
sns.heatmap(j_sq,annot=True)

plt.subplot(1,2,2)
(j_sq_w, j_w) = get_jaccard_sim(convMNlist.skeleton_id, convPSpartnersList.skeleton_id, weight='count')
sns.heatmap(j_sq_w,annot=True,)
```

```python
def plot_session_Ca_activity_list(cid, session, ROI, ytitle):
    plt.rcParams["figure.figsize"] = [6,2*len(cid)]
    fig, axs = plt.subplots(nrows=len(cid), ncols=2, sharex=True, sharey='row', squeeze=True)

    for idx,dummy in enumerate(cid):
        wL_Ca_trialMean = Ca_trial_means[session[idx]]['wL_trials'][int(ROI[idx]),:]
        bR_Ca_trialMean = Ca_trial_means[session[idx]]['bR_trials'][int(ROI[idx]),:]

        #ytitle = 'MN: ' + str(cid[idx]) + '\n' + session[idx][6:]
        #fig.suptitle(title[idx])

        axs[idx,0].set_ylabel(ytitle[idx])
        axs[idx,0].set_xlabel('left')    
        axs[idx,1].set_xlabel('right')

        plot_Ca_activity_blocks(wL_Ca_trialMean,axs[idx,0])
        plot_Ca_activity_blocks(bR_Ca_trialMean,axs[idx,1])
```

```python
def plot_Ca_activity_blocks(Ca_data_76,axs):
    #Define blocks
    cueBlockEarly = np.arange(13,27,1) #14:27; % 14 is running onset + 12 frames after
    cueBlockLate = np.arange(26,39,1) #27:39; % 12 frames before cue offset (frame 39)
    delayBlockEarly = np.arange(38,52,1) #39:52; % 39 is cue offset + 12 frames after
    delayTurnBlock = np.arange(51,65,1) #52:65; % 12 frames before end of trial (turn a certain amt)
    turnITIblock = np.arange(64,76,1) #65:76; % Trial end (reward given) an 12 frames after (dark, ITI)

    #Define colors
    cueEarlyColor = np.array([0, 161, 75])/255; 
    cueLateColor = np.array([255, 222, 23])/255;
    delayEarlyColor = np.array([237, 28, 36])/255;
    delayTurnColor = np.array([127, 63, 152])/255;
    turnITIcolor = np.array([33, 64, 154])/255;
    
    axs.plot(cueBlockEarly,Ca_data_76[cueBlockEarly],color=cueEarlyColor)
    axs.plot(cueBlockLate,Ca_data_76[cueBlockLate],color=cueLateColor)
    axs.plot(delayBlockEarly,Ca_data_76[delayBlockEarly],color=delayEarlyColor)
    axs.plot(delayTurnBlock,Ca_data_76[delayTurnBlock],color=delayTurnColor)
    axs.plot(turnITIblock,Ca_data_76[turnITIblock],color=turnITIcolor)

    return axs
```
