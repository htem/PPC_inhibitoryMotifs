```python
def scatter_and_fit(X, y, xlabel=None, ylabel=None, plot=True):
    X_ = X.reshape(-1, 1)
    y_ = y.reshape(-1, 1)
    reg = LinearRegression().fit(X_, y_)
    y_pred = reg.predict(X_)
    
    if plot == True:
        sns.scatterplot(X, y);
        if xlabel is not None:
            plt.xlabel(xlabel)
        if ylabel is not None:
            plt.ylabel(ylabel)
        plt.plot(X,y_pred)
        plt.title('R2 = %0.2f' % reg.score(X_,y_))
    return reg.score(X_,y_)

from scipy import stats
from sklearn.linear_model import LinearRegression
scatter_and_fit(myCN_df[myCN_df.source_type=='pyramidal']['RL_diff'].values,myCN_df[myCN_df.source_type=='pyramidal']['avg_psd_area'].values)
slope, intercept, r_value, p_value, std_err = stats.linregress(myCN_df[myCN_df.source_type=='pyramidal']['RL_diff'].values, myCN_df[myCN_df.source_type=='pyramidal']['avg_psd_area'].values)
print((slope, intercept, r_value, p_value, std_err ))
```

```python
import itertools

def get_synapses_between(source_skids, target_skids):
    connectors = pymaid.get_connectors_between(source_skids,target_skids)
    synapse_df = connectors[['source_neuron','target_neuron']]
    synapse_df["weight"] = 1
    synapse_df = synapse_df.rename(columns={"source_neuron": "source", "target_neuron": "target"})
    return synapse_df

def jaccard_sim_from_nx(G, source_skids):
    j = nx.jaccard_coefficient(G)
    j_df = pd.DataFrame(j)
    j_df = j_df.rename(columns={0: "source", 1: "target", 2: "weight"})
    j_G = nx.from_pandas_edgelist(j_df, edge_attr='weight')
    j_sq = nx.to_pandas_adjacency(j_G, nodelist = np.array(source_skids,dtype=int))
    #j_sim_sq_full = 1-j_sq#-np.eye(len(j_sq))
    #j_sim_sq = j_sim_sq_full.loc[np.array(source_skids,dtype=int),np.array(source_skids,dtype=int)]
    return j_sq

synapse_df = get_synapses_between(mySkids, int_psp_skel)
G=nx.from_pandas_edgelist(synapse_df)
j_sim_sq = jaccard_sim_from_nx(G, mySkids)
sns.heatmap(j_sim_sq)
```

```python
def bootstrap_synapses(synapse_df):
    synapse_bs = synapse_df.sample(frac=1, replace=True)
    return synapse_bs

def get_jaccard_sim(source_skids, target_skids):
    synapse_df = get_synapses_between(source_skids, target_skids)
    G = nx.from_pandas_edgelist(synapse_df)
    j_sim = jaccard_sim_from_nx(G, source_skids)
    return j_sim
    
def get_jaccard_bootstrap(source_skids, target_skids, num_samp, lower, upper):
    j_sim_bootstrap = np.zeros((len(source_skids),len(source_skids),num_samp))
    synapse_df = get_synapses_between(source_skids, target_skids)
    G = nx.from_pandas_edgelist(synapse_df)
    j_sim = jaccard_sim_from_nx(G, source_skids)
    for i in np.arange(num_samp):
        synapse_bs = bootstrap_synapses(synapse_df)
        G_bs = nx.from_pandas_edgelist(synapse_bs)
        j_sq_bs = jaccard_sim_from_nx(G_bs, source_skids)
        j_sim_bootstrap[:,:,i] = j_sq_bs
    j_sim_upper = np.percentile(j_sim_bootstrap, upper, axis=2)
    j_sim_lower = np.percentile(j_sim_bootstrap, lower, axis=2)
    
    return j_sim, j_sim_lower, j_sim_upper

(j_sim, j_sim_05, j_sim_95) = get_jaccard_bootstrap(mySkids, int_psp_skel, 10, 33, 68)
sns.set(rc={'figure.figsize':(20,6)})
plt.subplot(1,3,1)
sns.heatmap(j_sim_05)
plt.subplot(1,3,2)
sns.heatmap(j_sim)
plt.subplot(1,3,3)
sns.heatmap(j_sim_95)
```

```python
def plot_session_Ca_activity(cid, session, ROI, title):

    wL_Ca_trialMean = Ca_trial_means[session]['wL_trials'][ROI,:]
    bR_Ca_trialMean = Ca_trial_means[session]['bR_trials'][ROI,:]

    plt.rcParams["figure.figsize"] = [5,2.5]
    fig, axs = plt.subplots(nrows=1, ncols=2, sharex=True, sharey=True, squeeze=True)

    ytitle = 'MN: ' + str(cid) + '\n' + session
    fig.suptitle(title)

    axs[0].set_ylabel(ytitle)
    axs[0].set_xlabel('left')   
    axs[1].set_xlabel('right')

    for idx,trial_type in enumerate(trialTypes[:2]):
        plot_Ca_activity_blocks(wL_Ca_trialMean,axs[0])
        plot_Ca_activity_blocks(bR_Ca_trialMean,axs[1])
```

```python
def plot_session_Ca_activity_list(cid, session, ROI, ytitle):
    plt.rcParams["figure.figsize"] = [6,2*len(cid)]
    fig, axs = plt.subplots(nrows=len(cid), ncols=2, sharex=True, sharey='row', squeeze=True)

    for idx,dummy in enumerate(cid):
        wL_Ca_trialMean = Ca_trial_means[session[idx]]['wL_trials'][int(ROI[idx]),:]
        bR_Ca_trialMean = Ca_trial_means[session[idx]]['bR_trials'][int(ROI[idx]),:]

        #ytitle = 'MN: ' + str(cid[idx]) + '\n' + session[idx][6:]
        #fig.suptitle(title[idx])

        axs[idx,0].set_ylabel(ytitle[idx])
        axs[idx,0].set_xlabel('left')    
        axs[idx,1].set_xlabel('right')

        plot_Ca_activity_blocks(wL_Ca_trialMean,axs[idx,0])
        plot_Ca_activity_blocks(bR_Ca_trialMean,axs[idx,1])
```

```python
def plot_Ca_activity_blocks(Ca_data_76,axs):
    #Define blocks
    cueBlockEarly = np.arange(13,27,1) #14:27; % 14 is running onset + 12 frames after
    cueBlockLate = np.arange(26,39,1) #27:39; % 12 frames before cue offset (frame 39)
    delayBlockEarly = np.arange(38,52,1) #39:52; % 39 is cue offset + 12 frames after
    delayTurnBlock = np.arange(51,65,1) #52:65; % 12 frames before end of trial (turn a certain amt)
    turnITIblock = np.arange(64,76,1) #65:76; % Trial end (reward given) an 12 frames after (dark, ITI)

    #Define colors
    cueEarlyColor = np.array([0, 161, 75])/255; 
    cueLateColor = np.array([255, 222, 23])/255;
    delayEarlyColor = np.array([237, 28, 36])/255;
    delayTurnColor = np.array([127, 63, 152])/255;
    turnITIcolor = np.array([33, 64, 154])/255;
    
    axs.plot(cueBlockEarly,Ca_data_76[cueBlockEarly],color=cueEarlyColor)
    axs.plot(cueBlockLate,Ca_data_76[cueBlockLate],color=cueLateColor)
    axs.plot(delayBlockEarly,Ca_data_76[delayBlockEarly],color=delayEarlyColor)
    axs.plot(delayTurnBlock,Ca_data_76[delayTurnBlock],color=delayTurnColor)
    axs.plot(turnITIblock,Ca_data_76[turnITIblock],color=turnITIcolor)

    return axs
```
